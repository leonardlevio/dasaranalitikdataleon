```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/leonardlevio/dasaranalitikdataleon/refs/heads/main/high_popularity_spotify_data.csv')

popularity = data['track_popularity']

mean = popularity.mean()
mode = popularity.mode()
median = popularity.median()
variance = popularity.var()
std = popularity.std()
skew = popularity.skew()
kurt = popularity.kurtosis()

print('Popularity Informations:')
print(f'Mean: {mean:.2f}')
print(f'Median: {median:.2f}')
print(f'Modus: {mode[0]:.2f}')
print(f'Variance: {variance:.2f}')
print(f'Standard Deviation: {std:.2f}')
print(f'Skewness: {skew:.2f}')
print(f'Kurtosis: {kurt:.2f}')
plt.figure(figsize=(10,5))
plt.hist(popularity, bins=100, color='grey')
plt.axvline(mean, color='red', label='Mean')
plt.axvline(median, color='yellow', label='Median')
plt.axvline(mode[0], color='blue', label='Mode')
plt.ylabel('Tracks Count')
plt.xlabel('Popularity')
plt.legend()
plt.show()

data.describe()
```
